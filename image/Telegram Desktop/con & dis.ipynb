```python
class Math:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def add(self):
        return f"Addition of {self.x} and {self.y} is {self.x+self.x+self.y}"
    def sub(self):
        return  f"Addition of {self.x} and {self.y} is {self.x-self.x+self.y}"
    
obj4=Math(20,30)
print(obj4.add())
print(obj4.sub())
```

```python
class student:
    def __init__(self,Name,Add):
        self.Name = Name
        self.Add = Add
    def std_info(self):
        return f"student Name is {self.Name} and she is from {self.Add}"   
     
obj1=student('shraddha jadhav','pune')
obj1.std_info()     
```

```python
class Number:
    def __init__(self):
        pass
    def number(self,a,b):
        self.a = a
        self.b = b
        return f"value of a {a} and value of b {b}"
    
obj2=Number()
obj2.number(10,20)    
```

```python
class course:
    def __init__(self,c1,c2='UI'):
        self.c1 = c1
        self.c2 = c2
    def std_c(self):
        return f"{self.c1},{self.c2}"    
obj3 = course("python")
obj3.std_c()    
```

```python
class bank_details:
    def __init__(self,account_no):
        self.account_no = account_no
    def credit(self,amount_credited ):
        self.amount_credited = amount_credited
        return f'In account number {self.account_no} :Rs {amount_credited} is cedited'
    def debit(self,money):
        self.money = money
        return f'In account number {self.account_no} :Rs {money} is debited'
     
customer = bank_details('123456789')
print(customer.credit(10000))     
print(customer.debit(3000))      
```
