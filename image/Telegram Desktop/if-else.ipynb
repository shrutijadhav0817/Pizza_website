### 1. Check if a Number is Positive, Negative, or Zero

```python
num = int(input("Enter the number: "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
x = int(input("Enter a number: "))
if x > 0:
    print(f"{x} is positive")
elif x < 0:
    print(f"{x} is Negative")
else:
    print(f"{x} is Zero")
```

### 2. Determine if a Person is Eligible to Vote Based on Age

```python
age = int(input("Enter the age: "))
if age < 0 or age > 120:
    print("Invalid age")
elif age >= 18:
    print("Eligible to vote")
else:
    print("not eligible to vote")
```

### 3. Check if a Number is Divisible by Both 3 and 5

```python
num1 = int(input("Enter the number: "))
if num1 % 3 == 0 and num1 % 5 == 0:
    print("Divisible by Both 3 and 5")
elif num1 % 3 == 0:
    print("Divisible by 3")
elif num1 % 5 == 0:
    print("Divisible by 5")
else:
    print("Not divisible by 3 and 5")
    
```

### 4. Find the Largest of Three Numbers

```python
a = int(input("Enter the number a: "))
b = int(input("Enter the number b: "))
c = int(input("Enter the number c: "))

if a >= b and a >= c:
    print(f"{a} is Greater")
elif b >= c and b >= a:
    print(f"{b} is Greater")
else:
    print(f"{c} is Greater")
```

```python
a = int(input("Enter the number a: "))
b = int(input("Enter the number b: "))
c = int(input("Enter the number c: "))

largest = max(a, b, c)
print(f"{largest} is the Greatest")
```

### 5. Check if a Character is a Vowel or Consonant

```python
char1 = input("Enter the character to check vowel or consonant: ")
if len(char1) == 1 and char1.isalpha():
    if char1 in ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']:
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Invalid Input")
```

```python
char1 = input("Enter the character to check vowel or consonant: ")
list4 = ['a', 'e', 'i', 'o', 'u']
if len(char1) == 1 and char1.isalpha():
    if char1.lower() in list4:
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Invalid Input")
```

### 6. Determine if a Given Number is Even or Odd

```python
num2 = int(input("Enter the number: "))
if num2 % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### 7. Check if Two Numbers are Equal, Greater, or Lesser

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

if num1 == num2:
    print("Both numbers are equal")
elif num1 > num2:
    print(f"{num1} is greater than {num2}")
else:
    print(f"{num2} is greater than {num1}")
```

### 8. Determine the Grade Based on a Score (e.g., A, B, C, D, F)

```python
score = int(input("Enter the score: "))
if score < 0 or score > 100:
    print("Invalid score")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 35:
    print("D")
else:
    print("F")
```

### 9. Determine if a String Starts with a Vowel

```python
str1 = input("Enter the string: ")
list4 = ['a', 'e', 'i', 'o', 'u']
if len(str1) == 0:  # Check for empty string
    print("Empty string! No characters to check.")
elif str1[0].lower() in list4:
    print("String Starts with a Vowel")
else:
    print("String not Starts with a Vowel")
```

### 10. Check if a Year is a Leap Year

```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not leap year")
```

### 11. Implement a Simple Calculator Using if-elif-else Statements

```python
while True:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    operation = input("Enter operation (+, -, *, /): ")
    
    if operation.lower() == 'q':  # Exit the loop
        print("Calculator exiting...")
        break
    if operation == '+':
        print(f"The Addition of {num1} + {num2} = {num1 + num2}")
    elif operation == '-':
        print(f"The Subtraction of {num1} - {num2} = {num1 - num2}")
    elif operation == '*':
        print(f"The multiplication of {num1} * {num2} = {num1 * num2}")
    elif operation == '/':
        print(f"The division of {num1} / {num2} = {num1 / num2}")
    else:
        print("Invalid operation! Please enter +, -, *, /, or 'q' to quit.")
```

### 13. Check if a String is Empty or Not

### 15. Determine if a Person is a Child, Teen, or Adult Based on Age

```python
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age")
elif age <= 12:
    print("Child")
elif age <= 19:
    print("Teen")
else:
    print("Adult")
```
