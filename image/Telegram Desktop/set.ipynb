### 2. Find the Union of Two Sets

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
union_set = set2.union(set3)
union_set
```

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
union_set = set2 | set3
union_set
```

### 3. Find the Intersection of Two Sets

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
new_set = set2.intersection(set3)
new_set
```

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
new_set = set2 & set3
new_set
```

### 4. Find the Difference Between Two Sets

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
new_set = set2.difference(set3)
new_set
```

```python
set2 = {1,2,3,4,5}
set3 = {4,5,6,7,8,9,10}
new_set = set2 - set3
new_set
```

### 5. Check if a Set is a Subset of Another Set

```python
set4 = {1,2,3,4,5}
set5 = {4,5,6,7,8,9,10}
new_set = set4.issubset(set5)
new_set
```

```python
set4 = {1,2,3,4,5}
set5 = {1,2}
new_set = set5.issubset(set4)
new_set
```

```python
set4 = {1,2,3,4,5}
set5 = {1,2}
new_set = set5 <= set4
new_set
```

### 6. Check if Two Sets are Disjoint

```python
set1 = {1, 2, 3, 4}
set2 = {5, 6, 7, 8}
disjoint_Set = set1.isdisjoint(set2)
disjoint_Set
```

```python
set1 = {1, 2, 3, 4}
set2 = {5, 6, 7, 8, 4}
disjoint_Set = set2.isdisjoint(set1)
disjoint_Set
```

### 7. Remove an Element from a Set

```python
my_set = {10, 20, 30}
my_set.discard(40)
my_set
```

### 8. Add an Element to a Set

### 9. Find the Symmetric Difference Between Two Sets

```python
set9 = {1, 2, 3, 4}
set99 = {5, 6, 7, 8, 4}
sym_diff = set9.symmetric_difference(set99)
sym_diff
```

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
symmetric_diff = set_a ^ set_b
symmetric_diff
```

### 12. Check if Two Sets are Equal
