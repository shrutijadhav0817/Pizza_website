### 1. Print Numbers From 1 to 10 Using a While Loop

```python
num = 1
while num < 11:
    print(num, end=" ")
    num+=1
```

### 2. Calculate the Sum of Digits of a Given Number

```python
num2 = int(input("Enter the number: "))
sum=0
n = abs(num2)
while n > 0:
    rem = n%10
    sum+=rem
    n//=10
print(f"The sum of digit {num2} is {sum}")
```

### 3. Reverse a Number Using a While Loop

```python
num3 = int(input("Enter the number: "))
while num3 > 0:
    reverse = num3%10
    print(reverse,end="")
    num3//=10
```

```python
num3 = int(input("Enter the number: "))
reversed_num = 0
original_num = num3

while num3 > 0:
    digit = num3 % 10
    reversed_num = (reversed_num * 10) + digit
    num3 //= 10
print(f"The reverse of {original_num} is {reversed_num}")
```

### 4. Print a Countdown From 10 to 1

```python
n = 10
while n > 0:
    print(n, end=" ")
    n-=1
```

### 5. Find the Greatest Common Divisor (GCD) of Two Numbers

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
a = num1
b = num2
while b!=0:
    temp = b
    b = a%b
    a = temp
print(f"The GCD of {num1} and {num2} is {a}")
```

### 6. Keep Taking Input From the User Until They Enter 'exit'

```python
while True:
    enter_data = input("Enter the input: ")
    if enter_data=="Exit":
        break
```

### 7. Check if a Number is a Palindrome

```python
num7 = int(input("Enter the number: "))
n = num7
rev=0
while n != 0:
    rem = n%10
    rev = (rev * 10) + rem
    n//=10
if rev==num7:
    print("palindrome")
else:
    print("not a palindrome")
```

### 8. Generate the Fibonacci Sequence Until a Specified Number

```python
num8 = int(input("Enter the limit for the fibonacci sequence: "))
n = num8
a = 0
b = 1
while n!=0:
    print(a, end=" ")
    a,b=b,a+b
    n-=1
```

### 9. Calculate the Power of a Number Without Using the ** Operator

```python
base = int(input("Enter the base number: "))
exponent = int(input("Enter the exponent (positive integer): "))
result = 1
while exponent > 0:
    result = result * base  
    exponent = exponent - 1 
print(result)
```

### 11. Sum of Natural Numbers Until a Given Number

```python
num11 = int(input("Enter the limit: "))
n = 1
sum=0
while n <= num11:
    sum+=n
    n+=1
print(sum)
```

### 12. Implement a Simple Menu-Driven Program Using While Loop

```python
while True:
    a = int(input("Enter a: "))
    b = int(input("Enter b: "))
    operation = input("Enter which operation want to perform(+ - * / // % exit)")
    if operation=='+':
        print(f"{a} + {b} = {a+b}")
    elif operation=='-':
        print(f"{a} - {b} = {a-b}")
    elif operation=='*':
        print(f"{a} * {b} = {a*b}")
    elif operation=='/':
        print(f"{a} / {b} = {a/b}")
    elif operation=='//':
        print(f"{a} // {b} = {a//b}")
    elif operation=='%':
        print(f"{a} % {b} = {a%b}")
    else:
        print("Exiting code...")
        break
```

### 13. Simulate a Basic Password Check That Limits to 3 Attempts

```python
password1 = "1234"
attempt = 3
while attempt>=1:
    user_pass = input("Enter the password: ")
    if user_pass == 'exit':
        print("Exiting the code...")
        break
    elif user_pass == password1:
        print("login successfully")
        break
    else:
        attempt-=1
        print(f"Invalid Password. {attempt} attempts left.")
if attempt==0:
    print("You have reached your limit try again after 24 hours")
```

### 14. Keep Multiplying a Number by 2 Until It Becomes Greater Than 1000

```python
number = 1
while number <= 1000:
    print(f"Current number: {number}")
    number *= 2
print(f"Final number (greater than 1000): {number}")
```

### 15. Print the Digits of a Number in Reverse Order

```python
num = int(input("Enter the number: "))
if num == 0:
    print(0)
elif num < 0:
    print("Reversed number is not applicable for negative integers.")
else:
    n = num
    rev = 0
    while n != 0:
        rem = n % 10
        rev = (rev * 10) + rem
        n //= 10
    print(rev)
```

```python
number = int(input("Enter a number: "))
if number == 0:
    print(0)
else:
    while number > 0:
        digit = number % 10
        print(digit, end="")
        number //= 10
print()
```
