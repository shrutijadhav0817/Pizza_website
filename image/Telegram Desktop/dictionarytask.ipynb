```python

### 1. Access the Value Associated with a Key
my_dict = {"name": "Alice", "age": 30}
print(my_dict["name"])
print(my_dict.get("name"))
print(my_dict.get("age", 0)) # Syntax: dictionary_name.get(key, default_value)
print(my_dict.get("salary", "Unknown"))
print(my_dict.get("occupation")) 
```

```python
### 2. Check if a Key Exists in a Dictionary
my_dict = {"name": "Alex", "id": 101, "active": True, "roles": ["admin", "editor"]}
keys_view = my_dict.keys()
keys_view
```

```python
my_dict = {"name": "Alice", "age": 30, "city": "London"}
if "age" in my_dict:
    print("The 'age' key exists.")
if "occupation" not in my_dict:
    print("The 'occupation' key does not exist.")
```

```python
### 3. Iterate Over a Dictionary and Print All Key-Value Pairs
my_dict3 = {"name": "Alex", "id": 101, "active": True, "roles": ["admin", "editor"]}
key = my_dict3.keys()
value = my_dict3.values()
print(f"{key}\n{value}")
```

```python
my_dict3 = {"name": "Alex", "id": 101, "active": True, "roles": ["admin", "editor"]}
print("--- Key-Value Pairs using .items() ---")
for key, value in my_dict3.items():
    print(f"Key: {key} ==> Value: {value}")
```

```python
### 4. Merge Two Dictionaries
dict_a = {"a": 1, "b": 2, "c": 3}
dict_b = {"c": 99, "d": 4, "e": 5}
merged_dict = {**dict_a, **dict_b} # unpacking
merged_dict
```

```python
dict_1 = {"id": 1, "status": "active"}
dict_2 = {"status": "inactive", "timestamp": "2025-10-08"}
merged_dict = dict_1 | dict_2 # union operator
print(merged_dict)
```

```python
### 6. Find the Maximum and Minimum Values in a Dictionary
grades = {
    "Math": 92,
    "Science": 88,
    "History": 95,
    "Art": 85
}

max_value = max(grades.values())
min_value = min(grades.values())

print(f"Max Value: {max_value}")
print(f"Min Value: {min_value}")
```

```python
grades = {
    "Math": 92,
    "Science": 88,
    "History": 95,
    "Art": 85
}

max_value = max(grades.values())
max_key = max(grades, key=grades.get)

min_value = min(grades.values())
min_key = min(grades, key=grades.get)

print(f"Highest Grade: {max_key} ({max_value})")
print(f"Lowest Grade: {min_key} ({min_value})")
```

```python
### 7. Sort a Dictionary by Keys or Values
data = {"b": 30, "a": 40, "d": 10, "c": 20}
sorted_dict_by_key = dict(sorted(data.items())) 
print(f"Sorted Dictionary (by Key): {sorted_dict_by_key}")

sorted_dict_asc = dict(sorted(data.items(), key=lambda item: item[1], reverse=False))
print(f"Sorted Dictionary (by Value, ASC): {sorted_dict_asc}")

sorted_dict_asc = dict(sorted(data.items(), key=lambda item: item[0], reverse=True))
print(f"Sorted Dictionary (by key, DESC): {sorted_dict_asc}")
```

```python
### 10. Count the Frequency of Each Character in a String Using a Dictionary
text = "Python Programming"
freq = {}
for char in text:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print(freq)
```

```python
text1 = "programming is fun"
frequency = {}
for char in text1:
    frequency[char] = frequency.get(char, 0) + 1
print(frequency)
```

```python
from collections import Counter
text = "programming is fun"
frequency = Counter(text)
print(frequency)
print(type(frequency))
```

### 13. Reverse the Keys and Values in a Dictionary

```python
original_scores = {
    "Alice": 90, 
    "Bob": 85, 
    "Charlie": 90  # Duplicate value (90)
}
reversed_score = {v:k for k, v in original_scores.items()}
print(f"Original: {original_scores}")
print(f"Reversed: {reversed_score}")
```

```python
original_scores = {
    "Alice": 90, 
    "Bob": 85, 
    "Charlie": 90 
}
reversed_scores_loop = {}
for k, v in original_scores.items():
    reversed_scores_loop[v] = k
print(reversed_scores_loop)
```

### 14. Create a Dictionary Using Dictionary Comprehension

```python
grades = {"Alice": 85, "Bob": 92, "Charlie": 78, "David": 95}
passing_grades = {k:v for k, v in grades.items() if v >= 90}
passing_grades
```

```python
numbers = [1, 2, 3, 4, 5]
num_squares = {n: n*n for n in numbers}
num_squares
```
