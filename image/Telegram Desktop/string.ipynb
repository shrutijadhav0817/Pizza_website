```python
### 2. Check if a String is a Palindrome
str1 = input("Enter a string to check palindrome: ")
clean_str = str1.replace(" ", "").lower()

if clean_str == clean_str[::-1]:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
### 3. Count the Number of Vowels in a Stringstr5 = input("Enter the string: ")
count = 0
for i in str5:
    if i.lower() in ['a','e','i','o','u']:
        count += 1
print(f"The no. of Vowels in a string are: {count}")
```

```python
### 4. Find the Frequency of Characters in a String
string = input("Enter the string: ")
freq = {}
for char in string:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print("Character Frequency:")
for key, value in freq.items():
    print(f"{key} : {value}")
```

```python
### 5. Check if Two Strings are Anagrams
string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")

string1 = string1.replace(" ","").lower()
string2 = string2.replace(" ","").lower()

if sorted(string1) == sorted(string2):
    print("Strings are Anagrams")
else:
    print("Strings are not Anagrams")
```

```python
### 6. Remove Duplicate Characters from a String
str6 = input("Enter the string: ")
result = ""
for i in str6:
    if i not in result:
        result += i
print("String after removing duplicates:", result)
```

```python
### 7. Check if a String Contains Only Digits
str7 = input("Enter a string: ")
only_digits = True  # Assume all characters are digits

for ch in str7:
    if ch < '0' or ch > '9':  # Check if character is not a digit
        only_digits = False
        break

if only_digits and str7 != "":  # Ensure string is not empty
    print("The string contains only digits")
else:
    print("The string contains non-digit characters")
```

```python
str7 = input("Enter a string: ").strip()
if str7.isdigit():
    print("The string contains only digits")
else:
    print("The string contains non-digit characters")
```

```python
### 8. Convert the First Letter of Each Word to Uppercase
str8 = input("Enter a string: ")

# Method 1: Using title()
print("Converted string:", str8.title())

# Method 2: Without using title(), manual method
words = str8.split()  # Split string into words
#capitalized_words = [word[0].upper() + word[1:] if word else '' for word in words]
capitalized_words = []

for word in words:
    if word:  # Check if word is not empty
        capitalized_word = word[0].upper() + word[1:]
        capitalized_words.append(capitalized_word)
    else:
        capitalized_words.append('')  # Handle empty words
result = " ".join(capitalized_words)
print("Converted string:", result)
```
